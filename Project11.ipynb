```python


#WATER JUG
import heapq
def a_star_search(X, Y, D):
    def heuristic(x, y):
        return min(abs(D - x), abs(D - y))
    def is_goal(state):
        return D in state
    def get_neighbors(x, y):
        return [
            (X, y), (x, Y), (0, y), (x, 0),
            (x - min(x, Y - y), y + min(x, Y - y)),
            (x + min(y, X - x), y - min(y, X - x))]
    start = (0, 0)
    frontier = [(heuristic(0, 0), 0, start)]
    came_from = {start: None}
    while frontier:
        _, cost, current = heapq.heappop(frontier)
        if is_goal(current):
            path = []
            while current:
                path.append(current)
                current = came_from[current]
            return path[::-1]
        for neighbor in get_neighbors(*current):
            if neighbor not in came_from:
                heapq.heappush(frontier, (cost + 1 + heuristic(*neighbor), cost + 1, neighbor))
                came_from[neighbor] = current
    return None
X, Y, D = 7,3,5
solution = a_star_search(X, Y, D)
print("Solution found:" if solution else "No solution found.")
if solution:
    for state in solution:
        print(state)

#8 PUZZLE

import heapq

class PuzzleState:
    def __init__(self, board, goal, moves=0):
        self.board, self.goal, self.moves = board, goal, moves

    def eq(self, other):
        return self.board == other.board

    def lt(self, other):
        return (self.moves + self.heuristic()) < (other.moves + other.heuristic())

    def hash(self):
        return hash(str(self.board))

    def heuristic(self):
        return sum(abs((v-1)//3 - i) + abs((v-1)%3 - j) for i, r in enumerate(self.board) for j, v in enumerate(r) if v)

    def get_next_states(self):
        i, j = next((i, j) for i, r in enumerate(self.board) for j, v in enumerate(r) if not v)

def a_star(initial_state):
    open_set, closed_set = [], set()
    heapq.heappush(open_set, initial_state)
    while open_set:
        c = heapq.heappop(open_set)
        if c.board == c.goal:
            return c.moves
        closed_set.add(c)

    return -1

ib = [[2, 3, 8], [1, 6, 4], [7, 0, 5]]
gb = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
initial_state = PuzzleState(ib, gb)
print("Number of moves:", a_star(initial_state))
print("Initial Board:")
for r in ib:
    print(r)
print("Goal Board:")
for r in gb:
    print(r)

#COLOUR GRAPHING

class GraphColoringCsp():
    def _init_(self,vertices,edges,colors):
        self.vertices=vertices
        self.edges=edges
        self.colors=colors
    def is_complete(self,assignment):
        return len(assignment)==len(self.vertices)
    def is_consistent(self,vertex,color,assignment):
        return all(assignment.get(neighbor)!=color for neighbor in self.edges[vertex])
    def backtracking_search(self,assignment={}):
        if self.is_complete(assignment):
            return assignment
        vertex=next(v for v in self.vertices if v not in assignment)
        for color in self.colors:
            if self.is_consistent(vertex,color,assignment):
                result=self.backtracking_search({**assignment,vertex:color})
                if result:
                    return result
        return None
def main():
    vertices = ['n', 'nw', 'ne', 'm', 'mw', 'me', 's']
    edges = {
        'n': ['nw', 'ne'], 'nw': ['n', 'mw'], 'ne': ['n', 'me'],
        'm': ['nw', 'ne', 'mw', 'me'], 'mw': ['nw', 'm', 's'],
        'me': ['ne', 'm', 's'], 's': ['mw', 'me']
    }
    colors = ["green", "blue", "red"]
    solution=GraphColoringCsp(vertices,edges,colors).backtracking_search()
    print("Solution found" if solution else "No solution found.")
    print(solution)
if _name=="main_":
    main()

#GAME SEARCH

def print_board(board):
    for row in board:
        print(' | '.join(row))
        print('-' * 5)

def check_winner(board, player):
    return any(all(cell == player for cell in row) for row in board) or \
           any(all(row[i] == player for row in board) for i in range(3)) or \
           all(board[i][i] == player for i in range(3)) or \
           all(board[i][2-i] == player for i in range(3))

def minimax(board, is_max):
    if check_winner(board, 'O'): return 1
    if check_winner(board, 'X'): return -1
    if not any(' ' in row for row in board): return 0
    if is_max:
        best = -float('inf')
        for i, row in enumerate(board):
            for j, cell in enumerate(row):
                if cell == ' ':
                    board[i][j] = 'O'
                    best = max(best, minimax(board, False))
                    board[i][j] = ' '
        return best
    else:
        best = float('inf')
        for i, row in enumerate(board):
            for j, cell in enumerate(row):
                if cell == ' ':
                    board[i][j] = 'X'
                    best = min(best, minimax(board, True))
                    board[i][j] = ' '
        return best

def best_move(board):
    move = None
    best = -float('inf')
    for i, row in enumerate(board):
        for j, cell in enumerate(row):
            if cell == ' ':
                board[i][j] = 'O'
                score = minimax(board, False)
                board[i][j] = ' '
                if score > best:
                    best, move = score, (i, j)
    return move

board = [['X', 'O', 'X'], [' ', 'O', ' '], [' ', ' ', 'X']]
print_board(board)
move = best_move(board)
if move:
    board[move[0]][move[1]] = 'O'


    print_board(board)

#BAYES

pip install pgmpy
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import BayesianEstimator
from pgmpy.inference import VariableElimination
import pandas as pd
data = pd.DataFrame(data={'Rain': ['No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No'],
                          'TrafficJam': ['Yes', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No'],
                          'ArriveLate': ['Yes', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'No']})
model = BayesianNetwork([('Rain', 'TrafficJam'), ('TrafficJam', 'ArriveLate')])
model.fit(data, estimator=BayesianEstimator)
cpds = model.get_cpds()
for cpd in cpds:
    print("CPD for variable:", cpd.variable)
    print(cpd)
inference = VariableElimination(model)
query_result = inference.query(variables=['ArriveLate'], evidence={'Rain': 'Yes'})
print(query_result)

#VALUE-ITERATION:                                                                                                                                                      import numpy as np

def value_iteration(P, R, gamma=0.9, theta=1e-6):
    V = np.zeros(len(P))
    while True:
        delta = 0
        for s in range(len(P)):
            v = V[s]
            V[s] = max(sum(P[s, a, s_] * (R[s, a, s_] + gamma * V[s_]) for s_ in range(len(P))) for a in range(len(P[0])))
            delta = max(delta, abs(v - V[s]))
        if delta < theta:
            break
    policy = np.zeros(len(P), dtype=int)
    for s in range(len(P)):
        policy[s] = np.argmax([sum(P[s, a, s_] * (R[s, a, s_] + gamma * V[s_]) for s_ in range(len(P))) for a in range(len(P[0]))])
    return policy, V

P = np.array([[[0.7, 0.3], [0.4, 0.6]], [[0.5, 0.5], [0.6, 0.4]]])
R = np.array([[[5, 10], [-1, 2]], [[2, 1], [0, 3]]])
policy, V = value_iteration(P, R)
print("Policy:", policy)
print("Value:", V)

#POLICY-ITERATION                                                                                                                                                       import numpy as np

def policy_iteration(P, R, gamma=0.9):
    policy = np.zeros(len(P), dtype=int)
    V = np.zeros(len(P))
    while True:
        stable = True
        for s in range(len(P)):
            a = policy[s]
            new_v = sum(P[s, a, s_] * (R[s, a, s_] + gamma * V[s_]) for s_ in range(len(P)))
            if new_v != V[s]:
                V[s] = new_v
                stable = False
        if stable:
            break
        for s in range(len(P)):
            policy[s] = np.argmax([sum(P[s, a, s_] * (R[s, a, s_] + gamma * V[s_]) for s_ in range(len(P))) for a in range(len(P[0]))])
    return policy, V

P = np.array([[[0.7, 0.3], [0.4, 0.6]], [[0.5, 0.5], [0.6, 0.4]]])
R = np.array([[[5, 10], [-1, 2]], [[2, 1], [0, 3]]])
policy, V = policy_iteration(P, R)
print("Policy:", policy)
print("Value:", V)

#SUDOKU                                                                                                                                                                          def is_valid(board, row, col, num):
    for i in range(9):
        if board[row][i] == num or board[i][col] == num or board[3 * (row // 3) + i // 3][3 * (col // 3) + i % 3] == num:
            return False
    return True
def solve_sudoku(board):
    for row in range(9):
        for col in range(9):
            if board[row][col] == 0:
                for num in range(1, 10):
                    if is_valid(board, row, col, num):
                        board[row][col] = num
                        if solve_sudoku(board):
                            return True
                        board[row][col] = 0
                return False
    return True
sudoku_board = [[5, 3, 0, 0, 7, 0, 0, 0, 0], [6, 0, 0, 1, 9, 5, 0, 0, 0], [0, 9, 8, 0, 0, 0, 0, 6, 0],
                [8, 0, 0, 0, 6, 0, 0, 0, 3], [4, 0, 0, 8, 0, 3, 0, 0, 1], [7, 0, 0, 0, 2, 0, 0, 0, 6],
                [0, 6, 0, 0, 0, 0, 2, 8, 0], [0, 0, 0, 4, 1, 9, 0, 0, 5], [0, 0, 0, 0, 8, 0, 0, 7, 9]]
if solve_sudoku(sudoku_board):
    print("Sudoku Solved Successfully:")
    [print(row) for row in sudoku_board]
else:
    print("No solution exists.")
16 visits · 1 online
 Share
Vote: 0 0
  0  Save as PDF
© 2024 JustPaste.it
BlogAbout
```
